```python
import numpy as np
from tabulate import tabulate
```

Define functions that return sin and cos of x respectively.

```python
def sin(x):
    return np.sin(x)
```

```python
def cos(x):
    return np.cos(x)
```

```python
x = np.linspace(0, (2 * np.pi), 1000)
y = sin(x)
z = cos(x)
table_data = [(a, b, c) for a, b, c in zip(x, y, z)]
table_headers = ["sin(x)","x"]
table = tabulate(table_data, tablefmt = "grid", headers = table_headers, floatfmt = ".3f")
```

```python
for i in table_data[:10]:
    print(i)
```
