```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
```

```python
def trig_sin(a):
    b = np.sin(a)
    return b
def trig_cos(a):
    b = np.cos(a)
    return b
```

```python
figure, axes = plt.subplots(1, 2)

axes[0].plot(x, trig_sin(x))
axes[0].set_xlabel("x")
axes[0].set_ylabel("sin(x)")
axes[0].set_title(r'$sin(x)$')

axes[1].plot(x, trig_cos(x))
axes[1].set_xlabel("x")
axes[1].set_ylabel("cos(x)")
axes[1].set_title(r'$cos(x)$')

figure.subplots_adjust(wspace = 0.4)

axes[0].set_aspect('equal')
axes[1].set_aspect(np.pi)

plt.savefig("Journal2Session8.pdf", bbox_inches = 'tight', dpi = 600)
plt.show
```
