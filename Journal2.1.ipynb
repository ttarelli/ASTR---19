```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 1, 100)
```

```python
def exponent(a):
    b = np.exp(a)
    return b
```

```python
y = exponent(x)

plt.plot(x, y)
plt.xlabel("Time[milliseconds]")
plt.ylabel("Awesomeness")

plt.savefig('Journal2Session7Figure.pdf', bbox_inches = 'tight', dpi = 60)
plt.show()
```
